```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation

# Load your dataset
df = pd.read_csv('DataEU (1).csv')

# Assuming dateRep is in datetime format, if not, convert it
df['dateRep'] = pd.to_datetime(df['dateRep'])

# Calculate relative excess deaths (%)
df['RelativeExcessDeaths'] = df.groupby('countries')['deaths'].transform(lambda x: (x - x.median()) / x.median() * 100)

# Create an animated bar chart
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 8))
barplot = sns.barplot(x='countries', y='RelativeExcessDeaths', data=df, ax=ax, palette='viridis')
ax.set_title('Median Relative Excess Deaths (%) by Country Over Time', fontsize=16)
ax.set_xlabel('Countries', fontsize=14)
ax.set_ylabel('Relative Excess Deaths (%)', fontsize=14)
ax.tick_params(axis='x', rotation=45)
ax.legend().set_visible(False)

def update(frame):
    current_date = df['dateRep'].unique()[frame]
    data_subset = df[df['dateRep'] == current_date]
    barplot = sns.barplot(x='countries', y='RelativeExcessDeaths', data=data_subset, ax=ax, palette='viridis')
    ax.set_title(f'Median Relative Excess Deaths (%) on {current_date.strftime("%Y-%m-%d")}', fontsize=16)

# Animate the bar chart
animation = FuncAnimation(fig, update, frames=len(df['dateRep'].unique()), repeat=False)
plt.show()
```

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import contextily as ctx

# Load COVID-19 cases dataset
covid_data = pd.read_csv('DataEU (1).csv')

# Load NUTS3 geometries dataset for European countries
nuts3_gdf = gpd.read_file(r'C:\Users\35389\venv\nuts3_shapefile.shp')

# Merge COVID-19 data with NUTS3 geometries based on a common identifier
merged_data = pd.merge(nuts3_gdf, covid_data, left_on='nuts_id', right_on='nuts_id', how='left')

# Plot a map for each country
european_countries = merged_data['country'].unique()

for country in european_countries:
    # Filter data for the specific country
    country_data = merged_data[merged_data['country'] == country]

    # Plot the map
    fig, ax = plt.subplots(figsize=(10, 10))

    # Plot the NUTS3 boundaries
    nuts3_gdf.plot(ax=ax, edgecolor='black', facecolor='none')

    # Plot COVID-19 cases on the map
    country_data.plot(column='cases', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

    # Add a basemap for better context (requires internet connection)
    ctx.add_basemap(ax, crs=nuts3_gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.Mapnik)

    # Customize the plot
    plt.title(f'COVID-19 Cases in NUTS3 Regions - {country} (2021)')
    plt.xlabel('Longitude')
    plt.ylabel('Latitude')

    # Save or show the plot
    plt.savefig(f'covid_map_{country}.png')
    plt.show()
```
